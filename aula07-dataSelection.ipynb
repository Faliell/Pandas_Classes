```python
# pandas não suporta or and...
Brea = data.loc[(data.City == "Brea") & ((data["Price Range"] == "$$")|(data["Price Range"] == "$$$"))]
```

```python
# Chaining -  para separar linhas tem que clocar ()
good_res = (data
            .loc[data.City == "Fullerton"]
            .loc[data["Price Range"]== "$"]
            .loc[data["Rating"] >= 4.5]
            .loc[:, "Name": "Website"]
            .reset_index(drop=True)
            )
```
