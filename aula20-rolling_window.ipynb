```python
NY.head(2)
```

```python
NY.rolling(window="10D")["AvgTemperature"].mean()
```

```python
NY["rolling_mean_10D"] = NY.rolling(window="10D",min_periods=5)["AvgTemperature"].mean()
NY["rolling_mean_30D"] = NY.rolling(window="30D",min_periods=15)["AvgTemperature"].mean()
```

```python
NY
```

```python
results = NY.rolling(window="10D",min_periods=5)["AvgTemperature"].agg(["mean","median","max", "min", "std"])
```

```python
results
```

```python
NY[["AvgTemperature", "rolling_mean_10D",	"rolling_mean_30D"]].loc[NY.index <= "1995-12-31"].plot()
```

```python
# verificar se existe cidades com nomes duplicados
data["City"].nunique()
```

```python
((data["State"] + data["City"])).nunique()
```

```python
data["state_city"] = (data["State"] +"_" +data["City"])
```

```python
data
```

```python
us_cities_result = (
    data.groupby("state_city")
    .rolling(30, min_periods=15)["AvgTemperature"]
    .agg(["max","min","mean","std"])
    )
```

```python
us_cities_result
```

```python
us_cities_result.reset_index(inplace=True)
us_cities_result
```

```python
us_cities_result.loc[us_cities_result["state_city"] =="Wyoming_Cheyenne"].plot(x="Date",y = ["max","min"])
```

```python
# outra equação
NY_ewm = NY.ewm(alpha=0.2, adjust=False)["AvgTemperature"].mean()
NY_ewm
```
