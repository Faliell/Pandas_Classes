Achar duplicados e remover

```python
data.duplicated(subset=["Name", "City", "Street Address"])
# retorna uma mascara
```

```python
mask1 = data.duplicated(subset=["Name", "City", "Street Address"])
```

```python
data.loc[mask1].loc[data.Name == "Tierra Mia Coffee"]
```

```python
mask1 = data.duplicated(subset=["Name", "City", "Street Address"], keep="last")
#keep first, last - qual quero manter - False - ver os duplicados
data.loc[mask1].loc[data.Name == "Tierra Mia Coffee"]
```

```python
# ter o dado limpo -- ~ como not
data_clean =data[~mask1]
```

```python
data_clean.shape
```

```python
data_clean.loc[data_clean.Name == "Fire Wings"]
```

```python
data_clean2=data.drop_duplicates(subset=["Name", "City", "Street Address"], keep="first")
data_clean2.loc[data_clean2.Name == "Fire Wings"]
```
